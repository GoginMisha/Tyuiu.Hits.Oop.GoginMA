```polyglot-notebook
using System;
using System.Collections.Generic;

// Интерфейс для книг, которые можно брать
public interface IBorrowable
{
    string Borrow();
    string Return();
}

// Интерфейс для книг, которые можно читать
public interface IReadable
{
    string Read();
}

// Базовый класс Book
public class Book
{
    // Атрибуты
    public string Title { get; set; }
    public string Author { get; set; }
    public int YearOfPublication { get; set; }
    public int Pages { get; set; }
    public bool IsAvailable { get; set; }
    public string Language { get; set; }

    // Конструктор
    public Book(string title, string author, int yearOfPublication,  int pages, string language)
    {
        Title = title;
        Author = author;
        YearOfPublication = yearOfPublication;
        Pages = pages;
        IsAvailable = true;
        Language = language;
    }

    // Методы
    public virtual string GetInfo()
    {
        return $"Название: {Title}\nАвтор: {Author}\nГод публикации: {YearOfPublication}\nСтраницы: {Pages}\nЯзык: {Language}";
    }

    public string Reserve()
    {
        if (IsAvailable)
        {
            IsAvailable = false;
            return "Книга зарезервирована.";
        }
        return "Книга уже зарезервирована.";
    }

    public string ReturnBook()
    {
        IsAvailable = true;
        return "Книга возвращена.";
    }

    public string CheckAvailability()
    {
        return IsAvailable ? "Книга доступна." : "Книга недоступна.";
    }

    // Перегруженный метод
    public virtual string Read(int startPage, int endPage)
    {
        return $"Вы читаете страницы {startPage} по {endPage}.";
    }
}

// Производный класс TextBook (Учебник)
public class TextBook : Book, IBorrowable, IReadable
{
    public string Subject { get; set; }
    public string GradeLevel { get; set; }
    public string School { get; set; }

    // Конструктор
    public TextBook(string title, string author, int yearOfPublication, int pages, string language, string subject, string gradeLevel, string school)
        : base(title, author, yearOfPublication, pages, language)
    {
        Subject = subject;
        GradeLevel = gradeLevel;
        School = school;
    }

    // Переопределенный метод чтения
    public string Read()
    {
        return $"Вы читаете учебник по {Subject}.";
    }

    public string Borrow()
    {
        return "Вы взяли учебник.";
    }

    public string Return()
    {
        return "Вы вернули учебник.";
    }

    // Новый метод
    public string Study()
    {
        return $"Вы изучаете {Subject} на уровне {GradeLevel} в {School}.";
    }

    // Перегрузка метода
    public override string Read(int startPage, int endPage)
    {
        return $"Вы читаете учебник с {startPage} по {endPage} страницы, изучая {Subject}.";
    }
}

// Производный класс Fiction (Художественная литература)
public class Fiction : Book, IBorrowable
{
    public string Genre { get; set; }
    public string MainCharacter { get; set; }
    public bool IsSeries { get; set; }

    // Конструктор
    public Fiction(string title, string author, int yearOfPublication, int pages, string language, string genre, string mainCharacter, bool isSeries)
        : base(title, author, yearOfPublication, pages, language)
    {
        Genre = genre;
        MainCharacter = mainCharacter;
        IsSeries = isSeries;
    }

    // Переопределенный метод взятия книги
    public string Borrow()
    {
        return $"Вы взяли книгу жанра {Genre}.";
    }

    public string Return()
    {
        return "Вы вернули художественную книгу.";
    }

    // Новый метод
    public string DescribeMainCharacter()
    {
        return $"Главный персонаж книги — {MainCharacter}.";
    }

    // Перегрузка метода
    public override string Read(int startPage, int endPage)
    {
        return $"Вы читаете художественное произведение с {startPage} по {endPage} страницы.";
    }
}

// Производный класс ScientificLiterature (Научная литература)
public class ScientificLiterature : Book, IReadable
{
    public string FieldOfScience { get; set; }
    public bool IsPeerReviewed { get; set; }
    public string Institution { get; set; }

    // Конструктор
    public ScientificLiterature(string title, string author, int yearOfPublication, int pages, string language, string fieldOfScience, bool isPeerReviewed, string institution)
        : base(title, author, yearOfPublication, pages, language)
    {
        FieldOfScience = fieldOfScience;
        IsPeerReviewed = isPeerReviewed;
        Institution = institution;
    }

    // Переопределенный метод получения информации
    public override string GetInfo()
    {
        return $"{base.GetInfo()}\nНаучная область: {FieldOfScience}\nРецензируемая: {(IsPeerReviewed ? "Да" : "Нет")}\nУчреждение: {Institution}";
    }

    // Переопределенный метод чтения
    public string Read()
    {
        return $"Вы читаете научную литературу по {FieldOfScience}.";
    }

    // Новый метод
    public string ConductResearch()
    {
        return $"Вы проводите исследование в области {FieldOfScience} в {Institution}.";
    }
}

// Генерический класс для библиотеки
public class Library<T> where T : Book
{
    private List<T> books = new List<T>();

    public void AddBook(T book)
    {
        books.Add(book);
    }

    public void RemoveBook(T book)
    {
        books.Remove(book);
    }

    public void DisplayBooks()
    {
        foreach (var book in books)
        {
            Console.WriteLine(book.GetInfo());
            Console.WriteLine(book.CheckAvailability());
        }
    }
}

        TextBook mathTextbook = new TextBook("Алгебра 7 класс", "Мерзляк", 2016, 272, "Русский", "Математика", "Школьник", "Неполная средняя школа");
Fiction fantasyFiction = new Fiction("Тайна Третьей планеты", "Кир Булычёв", 1997, 256, "Русский", "Фантастика", "Алиса Селезнёва", true);
ScientificLiterature scienceBook = new ScientificLiterature("Астробиология", "Дэвид Кэтлинг", 2013, 160, "Английский", "Физика", true, "Оксфордский университет");


        // Создаем библиотеку учебников
        Library<TextBook> textbookLibrary = new Library<TextBook>();
        textbookLibrary.AddBook(mathTextbook);

        // Создаем библиотеку художественной литературы
        Library<Fiction> fictionLibrary = new Library<Fiction>();
        fictionLibrary.AddBook(fantasyFiction);

        // Создаем библиотеку научной литературы
        Library<ScientificLiterature> scienceLibrary = new Library<ScientificLiterature>();
        scienceLibrary.AddBook(scienceBook);

        // Демонстрация методов
        Console.WriteLine(mathTextbook.Study());
        Console.WriteLine(fantasyFiction.DescribeMainCharacter());
        Console.WriteLine(scienceBook.ConductResearch());

        // Перегрузка метода Read()
        Console.WriteLine(mathTextbook.Read(20, 40));
        Console.WriteLine(fantasyFiction.Read(60, 73));

        // Проверка библиотек
        Console.WriteLine("\nУчебники:");
        textbookLibrary.DisplayBooks();

        Console.WriteLine("\nХудожественная литература:");
        fictionLibrary.DisplayBooks();

        Console.WriteLine("\nНаучная литература:");
        scienceLibrary.DisplayBooks();
```
